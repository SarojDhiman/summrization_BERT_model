```
from summarizer import Summarizer
from pprint import pprint
```

```
with open("/content/25EASMarch-3369.txt", 'r') as file:
    data = file.read().replace('\n', '')
```

```
len(data)
```

```
data = data.replace("\ufeff", "") # here this \ufeff tell Zero Width No-Break Space we are replacing that term here
```

```
data[0:100]
```

```
model = Summarizer()
```

```
result = model(data, num_sentences=15, min_length=100) # here num_sen is the number of sentences
```

```
full = ''.join(result)
```

```
pprint(full)
```
